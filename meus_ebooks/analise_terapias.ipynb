```python
import PyPDF2
from collections import Counter
import pandas as pd
import re

arquivo_pdf = "Mindfulness.pdf"

try:
    with open(arquivo_pdf, "rb") as f:
        pdf = PyPDF2.PdfReader(f)
        total_paginas = len(pdf.pages)
        
        # Juntar o texto de TODAS as páginas para analisar
        texto_completo = ""
        for pagina in pdf.pages:
            texto_completo += pagina.extract_text() or ""
            
    # Limpeza básica do texto usando expressões regulares (re)
    palavras = re.findall(r'\b[a-zA-ZÁ-ÿ]{4,}\b', texto_completo.lower())
    
    # Lista de palavras comuns para ignorar (stop words)
    ignorar = {'para', 'como', 'mais', 'uma', 'com', 'este', 'esse', 'isso', 'pelo', 'pela', 'seus', 'suas'}
    palavras_filtradas = [p for p in palavras if p not in ignorar]
    
    # Contar a frequência das palavras
    contagem = Counter(palavras_filtradas)
    df_palavras = pd.DataFrame(contagem.most_common(10), columns=['Palavra', 'Frequência'])
    
    print(f"✅ Conexão estabelecida com sucesso!")
    print(f"📄 O e-book de Mindfulness tem {total_paginas} páginas analisadas.\n")
    print("📊 Top 10 palavras mais frequentes no seu livro:")
    print(df_palavras.to_string(index=False))
    
except Exception as e:
    print(f"❌ Erro ao ler o arquivo: {e}")
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Pegando o texto completo que já extraímos na célula anterior
# Vamos gerar a nuvem de palavras com um design limpo e profissional
nuvem = WordCloud(
    width=1200, 
    height=800, 
    background_color="#F3F0DF", # Cor creme elegante de fundo
    colormap="viridis",         # Paleta de cores moderna
    min_font_size=10
).generate(texto_completo)

# Plotando o gráfico na tela do VS Code
plt.figure(figsize=(12, 8), facecolor=None)
plt.imshow(nuvem, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=0)

# Salva a imagem automaticamente na sua pasta para você usar nas redes sociais!
plt.savefig("nuvem_mindfulness.png", dpi=300)
plt.show()
print("✅ Incrível! A imagem 'nuvem_mindfulness.png' foi gerada e salva na sua pasta!")
```

```python
import PyPDF2
from collections import Counter
import pandas as pd
import re
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Nome exato do arquivo do segundo e-book
arquivo_reiki = "Terapia_Integrativa_Reiki.pdf"

try:
    # 1. Extração de texto
    with open(arquivo_reiki, "rb") as f:
        pdf_reiki = PyPDF2.PdfReader(f)
        total_paginas_reiki = len(pdf_reiki.pages)
        
        texto_completo_reiki = ""
        for pagina in pdf_reiki.pages:
            texto_completo_reiki += pagina.extract_text() or ""
            
    # 2. Limpeza e filtragem de palavras
    palavras_reiki = re.findall(r'\b[a-zA-ZÁ-ÿ]{4,}\b', texto_completo_reiki.lower())
    
    ignorar = {
        'para', 'como', 'mais', 'uma', 'com', 'este', 'esse', 'isso', 
        'pelo', 'pela', 'seus', 'suas', 'sobre', 'está', 'também', 'quando',
        'pode', 'será', 'esta', 'estão', 'entre', 'onde', 'podos'
    }
    palavras_filtradas_reiki = [p for p in palavras_reiki if p not in ignorar]
    
    # 3. Contagem e exibição da tabela
    contagem_reiki = Counter(palavras_filtradas_reiki)
    df_reiki = pd.DataFrame(contagem_reiki.most_common(10), columns=['Palavra', 'Frequência'])
    
    print(f"✅ Conexão com o e-book de Reiki estabelecida!")
    print(f"📄 O e-book tem {total_paginas_reiki} páginas analisadas.\n")
    print("📊 Top 10 palavras mais frequentes no livro de Reiki:")
    print(df_reiki.to_string(index=False))
    print("-" * 50)
    
    # 4. Geração e salvamento da Nuvem de Palavras
    nuvem_reiki = WordCloud(
        width=1200, 
        height=800, 
        background_color="#005088", # Fundo azul profundo para contrastar com o Reiki
        colormap="autumn",          # Cores quentes (laranja/amarelo) ligadas à energia
        min_font_size=10
    ).generate(texto_completo_reiki)
    
    plt.figure(figsize=(12, 8))
    plt.imshow(nuvem_reiki, interpolation="bilinear")
    plt.axis("off")
    plt.tight_layout(pad=0)
    
    # Salva a imagem na pasta
    plt.savefig("nuvem_reiki.png", dpi=300)
    plt.show()
    print("🌅 Sucesso! A imagem 'nuvem_reiki.png' foi gerada e salva na sua pasta!")
    
except Exception as e:
    print(f"❌ Erro ao processar o arquivo de Reiki: {e}")
```
